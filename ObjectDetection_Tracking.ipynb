```python
from google.colab import files
uploaded = files.upload()  # yahi video file upload karo
video_path = list(uploaded.keys())[0]
print(video_path)
```

```python
from ultralytics import YOLO

# Model load karo (pehli dafa chalane pe automatically download hoga, thoda time lagega)
model = YOLO('yolov8n.pt')
```

```python
import cv2

cap = cv2.VideoCapture(video_path)

# Output video save karne ke liye setup
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter('output.mp4', fourcc, fps, (width, height))

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Detection + tracking dono ek saath
    results = model.track(frame, persist=True)

    # Boxes, labels, tracking IDs frame pe draw karo
    annotated_frame = results[0].plot()

    out.write(annotated_frame)
    frame_count += 1

cap.release()
out.release()
print(f"Done! {frame_count} frames processed.")
```
